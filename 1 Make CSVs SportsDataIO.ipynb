```python
url1 = "https://api.sportsdata.io/v3/cfb/scores/json/GamesByWeek/"
url3 = "?key=c65fd8000de24412b418590cb7fe86e1"
```

```python
# Initialize an empty DataFrame
df = pd.DataFrame()
```

```python
year = 2013
while year <= 2023:
    week = 1
    while week <= 16:
        url = url1 + str(year) + "/" + str(week) + url3
        response = requests.get(url, headers=headers)
        if response.status_code == 200:
          data = response.json()  # Assuming JSON response
#          df = df.append(pd.DataFrame(data), ignore_index=True)  # Append the data to the DataFrame
          df = pd.concat([df, pd.DataFrame(data)], ignore_index=True)  # Append the data to the DataFrame
        else:
          print("API request failed. Status code:", response.status_code)
        week += 1
    year += 1
```
